```python
import matplotlib.pyplot as plt
from matplotlib_venn import venn2

# Crear un diagrama de Venn con dos conjuntos
venn = venn2(subsets=(30, 20, 10), set_labels=('A', 'B'))

# Mostrar el gráfico
plt.title("Diagrama de Venn de dos conjuntos")
plt.show()
```

```python
from venn import venn

# Definir conjuntos
data = {
    "Set A": {1, 2, 3, 4},
    "Set B": {3, 4, 5, 6},
    "Set C": {1, 5, 7, 8}
}

# Crear el diagrama
venn(data)
plt.title("Diagrama de Venn con más de tres conjuntos")
plt.show()
```

```python
from venn import venn

# Definir conjuntos


c1 = {f'c1_{i}' for i in range(4)}
c2 = {f'c2_{i}' for i in range(631)}
c3 = {f'c3_{i}' for i in range(5)}
c4 = {f'c4_{i}' for i in range(8)}
c5 = {f'c5_{i}' for i in range(2)}
c6 = {f'c6_{i}' for i in range(4)}
c7 = {f'c7_{i}' for i in range(3)}


set_a = set()
set_b = set()
set_c = set()
data = {
    "Rejected no enough time": set_a.union(c1).union(c2).union(c3).union(c4),
    "Rejected no enough resources": set_b.union(c2).union(c3).union(c5).union(c6),
    "Rejected all servers out": set_c.union(c3).union(c6).union(c7).union(c4)
}

# Crear el diagrama
venn(data)
plt.title("Servers assigned not completed")
plt.show()
```
