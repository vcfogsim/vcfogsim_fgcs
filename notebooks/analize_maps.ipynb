```python
import osmnx as ox

# Define the city
place = "Barcelona, Spain"

# Download the street network
G = ox.graph_from_place(place, network_type='drive')

# Plot the graph
fig, ax = ox.plot_graph(G, node_size=0, edge_color='white', bgcolor='black', edge_linewidth=0.5)
```
