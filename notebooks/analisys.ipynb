```python
import os
import matplotlib.pyplot as plt

base = ".."
inputs_folder = "inputs"
outputs_folder = "outputs"
```

```python
filenames = ["users.txt"]

titels = ["random", "nearest", "k-nearest", "predictive"]

for i,fil in enumerate(filenames):
# --- Paso 1: Leer el archivo ---
    filename = fil

    # --- Paso 2: Inicializar el diccionario ---
    user_dict = {}

    # --- Paso 3: Procesar cada línea del archivo ---
    with open(os.path.join(base,filename), 'r') as file:
        for line in file:
            parts = line.strip().split()
            # Extraer los valores usando los índices conocidos
            id_val = int(parts[0].split(":")[1])
            proc_val = float(parts[2])
            wait_val = float(parts[4])
            assign_val = float(parts[6])

            # Guardar en el diccionario
            user_dict[id_val] = [proc_val, wait_val, assign_val]

    # --- Paso 4: Mostrar el diccionario resultante ---
    #print(user_dict)

    # Extract data from user_dict
    processing_times = [values[0] for values in user_dict.values()]
    waiting_times = [values[1] for values in user_dict.values()]
    assignment_times = [values[2] for values in user_dict.values()]


    print(round(sum(processing_times),3),round(sum(waiting_times),3),round(sum(assignment_times),3))

    # Combine data for the boxplot
    # data = [processing_times, waiting_times, assignment_times]
    #
    # # Create the boxplot
    # plt.boxplot(data, tick_labels=["Processing Times", "Waiting Times", "Assignment Times"])
    # plt.title(f"Jobs Time {titels[i]}")
    # plt.ylabel("Time")
    # plt.savefig(os.path.join(base,outputs_folder,f"{titels[i]}.png"))
    # plt.show()
```
