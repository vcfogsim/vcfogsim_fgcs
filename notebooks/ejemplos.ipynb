```python
  from itertools import chain,accumulate

L = [20,100,10]
L =  [v*[(i % 2)] for i,v in enumerate(L)]
L = list(chain.from_iterable(L))
L[21:40]=[0]*(40-21)
L = [[e] for e in L]
LL = []
for e in L:
    if LL==[]:
        LL.append(e)
    elif LL[-1][0]==e[0]:
        LL[-1].append(e[0])
    else:
        LL.append(e)
L = [len(l) for l in LL]
print(L)
```
